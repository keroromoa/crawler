```python
import re
import time

import requests

DATA_PATH = './petforum_dog_chat'
DATA_PATH2 = './petforum_dog_chat_thread'
CRAWL_PAGE_CNT = 10
#URL_TEMPLATE = "https://www.ptt.cc/bbs/Gossiping/index{}.html"
URL_TEMPLATE = "https://www.petforums.co.uk/forums/dog-chat.2/page-{}"
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:45.0) Gecko/20100101 Firefox/45.0'
}
HOST = "https://www.petforums.co.uk"

def get_w_cookie(url):
    resp = requests.get(url)
    return resp
    
def get_list_page(url):
    """GET列表頁，取得內文頁的連結們

    Parameters
    ----------
    url : str
        PTT 列表頁URL

    Returns
    -------
    list
       PTT內文頁的links
    """
    resp = get_w_cookie(url)
    links = re.findall('''<a href="(threads/.+/)"
					title=""
					class="PreviewTooltip"
					data-previewUrl="threads/.+">''', resp.text)
    print(links)
    detail_page_links = ["%s/%s"%(HOST, link) for link in links]
    return detail_page_links

def get_list_page2(url):
    """GET列表頁，取得內文頁的連結們

    Parameters
    ----------
    url : str
        PTT 列表頁URL

    Returns
    -------
    list
       PTT內文頁的links
    """
    resp = get_w_cookie(url)
    links = re.findall('''<a href="(members/.+.\d+/)" class="avatar Av\d+m" data-avatarhtml="true">''', resp.text)
    print(links)
    detail_page_links = ["%s/%s"%(HOST, link) for link in links]
    return detail_page_links

def dump_page(url):
    """GET url的HTML並且寫到檔案裡

    Parameters
    ----------
    url : str
        PTT 內文頁URL

    Returns
    -------
    str
       儲存的檔案名稱
    """
    filename = "_".join(re.findall('members/(.+).(\d+)/', url)[0]) + '.html'
    resp = get_w_cookie(url)

    with open(DATA_PATH + '/' + filename, 'w') as f:
        f.write(resp.text)
    return resp

def dump_page2(url):
    """GET url的HTML並且寫到檔案裡

    Parameters
    ----------
    url : str
        PTT 內文頁URL

    Returns
    -------
    str
       儲存的檔案名稱
    """
    filename = "_".join(re.findall('(threads)/(.+)/', url)[0]) + '.html'
    resp = get_w_cookie(url)

    with open(DATA_PATH2 + '/' + filename, 'w') as f:
        f.write(resp.text)
    return resp
```

```python
for pg in range(40,5072):
    url = URL_TEMPLATE.format(pg)
    for link in get_list_page(url):
        link_pg = 1
        while True:
            """
            內文頁換頁邏輯
            """
            link_url = link + "page-%s"%link_pg
            resp = dump_page2(link_url)
            for memberinfo in get_list_page2(link_url):
                print(memberinfo)
                dump_page(memberinfo)
            
            """
            用canonical address來做判斷是否爬到最後一頁
            """
            canonical = re.findall('<link rel="canonical" href="(.+)" />', resp.text)[0]
            if link_pg == 1: #第一頁的特別處理，拿掉p=1
                link_url = link_url[:-6]
            if canonical != link_url:
                break
            link_pg += 1
```

```python
from bs4 import BeautifulSoup
import re
import requests
import json


#!/usr/bin/python
# -*- coding: utf-8 -*-
from os import listdir
from os.path import isfile, isdir, join

# 指定要列出所有檔案的目錄
mypath = "./petforum_dog_chat"

# 取得所有檔案與子目錄名稱
files = listdir(mypath)

# 以迴圈處理

signature = []
for f in files:
  # 產生檔案的絕對路徑
    fullpath = join(mypath, f)
  # 判斷 fullpath 是檔案還是目錄
    with open(fullpath) as file:
        f_text = file.read()
        soup = BeautifulSoup(f_text, 'lxml') # Beautifulsoup Object
        try:
            soupinfo = soup.select_one('#info')
            soupinfotext = ''.join(str(e) for e in soupinfo)
            soupinfolist = BeautifulSoup(soupinfotext, 'lxml')
            soupsignature = soupinfolist.select_one('.signature').text.replace('\n', '  ').replace('\u200b', '  ')
            signature.append(soupsignature)
        except TypeError:
            continue
        except AttributeError:
            signature.append('null')
print(signature)
print(len(signature))
```
